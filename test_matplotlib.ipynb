# Using % inline and imports 

```python
%matplotlib  inline  
#note the plt.show needs the %mat inline, but since the image is being saved 
#you no longer need the plt.show 
import numpy as np #when we plot it will show us in the next line of code
import matplotlib.pyplot as plt
```

## Plot a function 
* create an array 
* create another array that passes the values of the first 
* plot those arrays 
* define labels and legend etc. through matplotlib tech. 

## Question: 
* how does this plt. know to pass a line? 
* we have an array of values 
* we pass it through anther math function that transforms the values
* would it not plot points vs. a curve? 


```python
x = np.arange(0,5,0.1) #exclusive (start,end,increment)
y = np.sin(x) #exclusive? 
z = np.cos(x)

plt.plot(x,y, label = 'sin(x)')
plt.plot(x,z, label = 'cos(x)')

plt.xlabel('x')
plt.ylabel('y = sin(x), cos(x)')

plt.legend()
#plt.show()


plt.savefig('sinx_cosx.png',bbox_inches='tight',dpi=600)
#bbox bounding type is little lightspace
#this saves to a file 

```

### Exclusive and Inclusive 
* this is a mathematical describtion 
> * (0,5) is exclusive 
> * [0,5] is inclusive
